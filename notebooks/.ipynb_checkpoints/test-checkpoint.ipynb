```python
def filter_min_count(df, column, min_count):
    freq_df = dt.group_by(df, [column])
    min_count_values = freq_df[freq_df['count'] >= min_count][column].values
    return df[df[column].isin(min_count_values)]


def group_to_flatten_list(df: pd.DataFrame, group_column: str, flatten_column: str):
    result = {}
    for _, row in df.iterrows():
        key, value  = row[group_column], row[flatten_column]
        if key in result:
            result[key].append(value)
        else:
            result[key] = [value]

    return result


def group_to_flatten_set(df: pd.DataFrame, group_column: str, flatten_column: str):
    result = {}
    for _, row in df.iterrows():
        key, value  = row[group_column], row[flatten_column]
        if key in result:
            result[key].add(value)
        else:
            result[key] = set([value])

    return result


def sliding_window(sequences: list[list[int]], window_size: int):
    result = set()
    for sequence in sequences:
        for seq_element in np.lib.stride_tricks.sliding_window_view(sequence, window_size):
            result.add(tuple(seq_element))

    return [list(r) for r in result]


def features_target_split(sequences):
    features, targets = [], []
    for seq in sequences:
        seq_len = len(seq)
        features.append(seq[:seq_len-1])
        targets.append(seq[seq_len-1])

    return features, targets

seq_to_emb_seq = lambda seq, movie_embeddings: torch.stack([torch.tensor(movie_embeddings[str(int(s))]) for s in seq])

def to_tensor(obs, device, columns): 
    data = obs[columns]
    if type(data) == pd.DataFrame:
        data = data.values
    return torch.tensor(data).to(device)

features_fn = lambda obs, device: to_tensor(obs, device, ['user_id', 'movie_id'])
target_fn   = lambda obs, device: to_tensor(obs, device, ['user_movie_rating'])


def to_one_hot_vector(indexes, size):
    matrix = torch.zeros((indexes.shape[0], size))
    for row, idx in enumerate(indexes):
        matrix[row, idx] = 1
    return matrix


def to_dict(df, key_colum, value_column, key_cast = lambda k: k, value_cast = lambda v: v):
    result = {}
    for _, row in df[[key_colum, value_column]].iterrows():
        result[key_cast(row[key_colum])] = value_cast(row[value_column])
    return result
```

```python
def init_embedding_weights(embedding, weight_matrix, requires_grad = True):
    embedding.weights = torch.nn.Parameter(weight_matrix)
    embedding.weight.requires_grad = requires_grad
    return embedding

class Model(torch.nn.Module):
    def __init__(self, embedding_dim, output_size, hidden_dim, n_layers, device, dropout=0.2):
        super(Model, self).__init__()

        self.device     = device

        # Defining some parameters
        self.hidden_dim = hidden_dim
        self.n_layers   = n_layers

        self.embedding = torch.nn.Embedding(output_size, embedding_dim).to(self.device)
        
        #Defining the layers
        # RNN Layer
        self.rnn = torch.nn.GRU(
            embedding_dim, 
            hidden_dim, 
            n_layers, 
            batch_first = True,
            dropout     = dropout
        ).to(self.device)

        # Fully connected layer
        self.fc      = torch.nn.Linear(hidden_dim, output_size).to(self.device)
        self.softmax = torch.nn.Softmax(dim=1).to(self.device)

    def forward(self, x):
        batch_size = x.size(0)

        x_embed = self.embedding(x.to(self.device))

        # Initializing hidden state for first input using method defined below
        hidden      = self.init_hidden(batch_size)

        # Passing in the input and hidden state into the model and obtaining outputs
        out, hidden = self.rnn(x_embed, hidden)

        # Get all last outputs for all batchs
        out = out[-1]

        # Reshaping the outputs such that it can be fit into the fully connected layer
        out = out.contiguous().view(-1, self.hidden_dim)
        out = self.fc(out)
        out = self.softmax(out)

        return out

    def init(self, embedding_weight_matrix, batch_size):
        self.init_hidden(batch_size)
        self.embedding = init_embedding_weights(
            self.embedding, 
            embedding_weight_matrix, 
            requires_grad = True
        )
        
    def init_hidden(self, batch_size):
        # This method generates the first hidden state of zeros which we'll use in the forward pass
        # We'll send the tensor holding the hidden state to the device we specified earlier as well
        return torch.zeros(self.n_layers, batch_size, self.hidden_dim).to(self.device)
```

Create a movies sequences dataset from previous dataset:

```python
if CREATE_DATASET:
    interactions_df = filter_min_count(
        dataset.data,
        column    = 'user_id',
        min_count = MIN_INTERACTIONS
    )
    # Update movies sequenses for users with more than MIN_INTERACTIONS...
    sequencer = dt.Sequencer('movie_id', 'movie_min_int_seq')
    df        = sequencer.perform(interactions_df)

    
    movie_seqs_by_user_id = group_to_flatten_list(
        df,
        group_column   = 'user_id',
        flatten_column = 'movie_min_int_seq'
    )

    raw_movie_seq_time_series = sliding_window(
        list(movie_seqs_by_user_id.values()),
        window_size = MIN_INTERACTIONS
    )

    movie_seq_time_series, next_movie_seq = features_target_split(raw_movie_seq_time_series)

    train_set = {
        'movie_seq_time_series' : movie_seq_time_series,
        'next_movie_seq'        : next_movie_seq,
        'movie_id_by_seq'       : to_dict(df, 'movie_min_int_seq', 'movie_id',  key_cast=int, value_cast=int),
        'movie_seq_by_id'       : to_dict(df, 'movie_id',  'movie_min_int_seq', key_cast=int, value_cast=int)
    }

    ut.JSON.save(USER_MOVIE_SEQUENCES_PATH, train_set)
```

Load sequences train set and embeddings:

```python
embeddings_mean_by_movie_id = ut.JSON.load(MOVIE_MEAN_EMBEDDINGS_PATH)
train_set                   = ut.JSON.load(USER_MOVIE_SEQUENCES_PATH)
```

```python
embedding_dim = 768   # The number of expected features in the input x
hidden_size   = 1000  # The number of features in the hidden state h
num_layers    = 2     # Number of recurrent layers. E.g., setting num_layers=2
                    # would mean stacking two GRUs together to form a stacked 
                    # GRU, with the second GRU taking in outputs of the first
                    # GRU and computing the final results. Default: 1
num_classes   = max(train_set['next_movie_seq'])+1
batch_size    = 8

# Define hyperparameters
n_epochs      = 10
lr            = 0.00001
dropout       = 0.5
```

```python
embedding_weights = []

for seq in range(num_classes):
    movie_id         = train_set['movie_id_by_seq'][str(seq)]
    embeddind_vector = embeddings_mean_by_movie_id[str(movie_id)]
    embedding_weights.append(embeddind_vector)

embedding_weights_matrix = torch.tensor(embedding_weights)
embedding_weights_matrix.shape
```

```python
features = torch.tensor(train_set['movie_seq_time_series'])
targets  = torch.tensor(train_set['next_movie_seq'])

total_samples = features.shape[0]

features.shape, targets.shape
```

```python
# Instantiate the model with hyperparameters
model = Model(
    embedding_dim = embedding_dim, 
    output_size   = num_classes, 
    hidden_dim    = hidden_size, 
    n_layers      = num_layers,
    device        = pu.get_device(),
    dropout       = dropout
)

model.init(embedding_weights_matrix, batch_size)

# Define Loss, Optimizer
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=lr)

model
```

```python
sequences_dataset     = ds.SimpleDataset(features,  targets)
sequences_data_loader = sequences_dataset.as_data_loader(batch_size = batch_size, shuffle = True)
```

```python
# We'll use this helper to compute accuracy
def binary_accuracy(preds, y):
    #round predictions to the closest integer
    rounded_preds = torch.round(preds)
    
    correct = (rounded_preds == y).float() 
    acc = correct.sum() / len(correct)
    return acc
```

```python
from IPython.display import clear_output

feature_transform = lambda x: torch.permute(x, (1, 0))

class ASCIProgressBar:
    def __init__(self, total, update_steps = 100):
        self.update_steps = update_steps
        self.total        = total
        self.bar          = {
            "|": "/",
            "/": "-",
            "-": "\\",
            "\\": "|"
        }
        self.reset()

    def reset(self):
        self.count      = 0
        self.bar_count  = 0
        self.last_state = '|'
    
    def update(self, times, message=''):
        self.count += times
        if self.bar_count % 100 == 0:
            clear_output(wait=True)
            print(f'Progress {self.last_state} {int((self.count/self.total)*100)} %', message, end='')
            self.last_state = self.bar[ self.last_state]
        self.bar_count +=1
```

```python
def train(model, sequences_data_loader, n_epochs, total_samples):
    progress_bar  = ASCIProgressBar(
        total        = total_samples,
        update_steps = 100
    )

    # Training Run
    for epoch in range(1, n_epochs + 1):
        epoch_loss = 0.0    
        progress_bar.reset()
        for features, target in sequences_data_loader:
            optimizer.zero_grad()

            input  = feature_transform(features)

            y_pred = model(input)

            # print(y_pred.shape, target.shape)
            y_true = to_one_hot_vector(target, num_classes).to(pu.get_device())

            loss = criterion(y_pred, y_true)

            # Does backpropagation and calculates gradients
            loss.backward()

            # Updates the weights accordingly
            optimizer.step()

            epoch_loss += loss.item()

            current_count = 0.0001 if progress_bar.count ==0 else progress_bar.count

            progress_bar.update(
                times   = input.shape[1], 
                message = f'- Epoch: {epoch}/{n_epochs} - Loss: {epoch_loss/current_count:.4f}'
            )
```

```python
train(model, sequences_data_loader, n_epochs, total_samples)
```
