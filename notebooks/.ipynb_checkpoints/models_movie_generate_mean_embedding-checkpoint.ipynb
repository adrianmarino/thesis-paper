## Build consolidated embedding table

```python
def load(name): return pd.read_json(f'../datasets/movie_{name}_embedding_bert.json')
```

```python
embedding_tables  = [load(name) for name in ['genres', 'overview', 'tags', 'title']]
```

```python
def to_embeddding_matrix(df, column):
    shape = df[column].values.shape
    return np.concatenate(df[column].apply(np.array).values, axis=0).reshape(shape[0], -1)

to_embedding_df = lambda df, column: pd.DataFrame(to_embeddding_matrix(df, column))
```

```python
df= ut.merge_all(embedding_tables, 'id')
df = df[['id', 'release_year', 'imdb_id', 'title', 'genres_embedding', 'overview_embedding', 'tags_embedding', 'title_embedding']]
```

```python
embedding_columns = ['genres_embedding', 'overview_embedding', 'tags_embedding', 'title_embedding']

def list_columns_mean(df, columns):
    mean_embeddings = []
    for index, row in df.iterrows():
        embeddings = [row[c] for c in embedding_columns]
        matrix = np.stack(embeddings)
        mean_embeddings.append(matrix.mean(0))    

    return mean_embeddings
```
